```python
# 183. Customers Who Never Order
# https://leetcode.com/problems/customers-who-never-order/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def find_customers_183(customers: pd.DataFrame, orders: pd.DataFrame) -> pd.DataFrame:
    # Merge the two DataFrames on the customerId column
    joined_df = customers.merge(orders, left_on="id", right_on="customerId", how="left")

    # Return the customers who have not ordered anything
    return joined_df.query("id_y.isnull()")[["name"]].rename(
        columns={"name": "Customers"}
    )
```

```python
# 1148. Article Views I
# https://leetcode.com/problems/article-views-i/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def article_views_1148_1(views: pd.DataFrame) -> pd.DataFrame:
    filtered_df = views.query("author_id == viewer_id")[["author_id"]].rename(
        columns={"author_id": "id"}
    )
    # Removed duplicates & Sorted in ascending order
    return filtered_df.drop_duplicates().sort_values(by=["id"], ascending=True)
```

```python
# 1873. Calculate Special Bonus
# https://leetcode.com/problems/calculate-special-bonus/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def calculate_special_bonus_1873(employees: pd.DataFrame) -> pd.DataFrame:
    def g(df):
        df.loc[
            (df["employee_id"] % 2 == 0) | df["name"].str.startswith("M"), "salary"
        ] = 0
        return df.sort_values(["employee_id"], ascending=True)

    return g(employees)[["employee_id", "salary"]].rename(columns={"salary": "bonus"})
```

```python
# 1667. Fix Names in a table
# https://leetcode.com/problems/fix-names-in-a-table/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def fix_names_1667_1(users: pd.DataFrame) -> pd.DataFrame:
    def proper_case(df):
        df['name'] = df['name'].str.capitalize()
        return df.sort_values(['user_id'],ascending=True)
    return proper_case(users)
```

```python
# 1517. Find Users with Valid Emails
# https://leetcode.com/problems/find-users-with-valid-e-mails/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def valid_emails_1517_1(users: pd.DataFrame) -> pd.DataFrame:
    def valid_email_check(df):
        return df[df["mail"].str.match(r"^[a-zA-Z][a-zA-Z0-9_.-]*@leetcode\.com$")]

    return valid_email_check(users.copy())
```

```python
# https://leetcode.com/problems/find-users-with-valid-e-mails/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def find_patients_1527_2(patients: pd.DataFrame) -> pd.DataFrame:
    def diabetes_check(df):
        return df[
            df["conditions"].str.match(r"^DIAB1.*$")
            | df["conditions"].str.match(r"^.*\ DIAB1.*$")
        ]

    return diabetes_check(patients)
```

```python
# 177. Nth Highest Salary
# https://leetcode.com/problems/nth-highest-salary/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def nth_highest_salary_177(employee: pd.DataFrame, N: int) -> pd.DataFrame:
    if employee["salary"].nunique() < N:
        return pd.DataFrame(columns=["salary"])
    else:
        return (
            employee["salary"]
            .drop_duplicates()
            .sort_values(ascending=False)
            .reset_index()
            .loc[[N - 1], ["salary"]]
        )
```

```python
# 176. Second Highest Salary
# https://leetcode.com/problems/second-highest-salary/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def second_highest_salary_176(employee: pd.DataFrame) -> pd.DataFrame:
    if employee["salary"].nunique() < 2:
        return pd.DataFrame({"SecondHighestSalary": [None]})
    else:
        return (
            employee["salary"]
            .drop_duplicates()
            .sort_values(ascending=False)
            .reset_index()
            .loc[[1], ["salary"]]
            .rename(columns={"salary": "SecondHighestSalary"})
        )
```

```python
# 178. Rank Scores
# https://leetcode.com/problems/department-highest-salary/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def order_scores_178(scores: pd.DataFrame) -> pd.DataFrame:
    scores["rank"] = scores["score"].rank(method="dense", ascending=False)
    return scores.drop("id", axis=1).sort_values(by="score", ascending=False)
```

```python
# 1173.Immediate Food Delivery 1
# https://leetcode.com/problems/immediate-food-delivery-i/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def food_delivery_1173(delivery: pd.DataFrame) -> pd.DataFrame:
    immediate_percentage = (
        delivery[
            delivery["order_date"] == delivery["customer_pref_delivery_date"]
        ].shape[0]
        / delivery.shape[0]
    ) * 100
    return pd.DataFrame({"immediate_percentage": [round(immediate_percentage, 2)]})
```

```python
# 1741. Find Total Time Spent by Each Employee
# https://leetcode.com/problems/find-total-time-spent-by-each-employee/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def total_time_1741(employees: pd.DataFrame) -> pd.DataFrame:
    employees["time_spent"] = employees["out_time"] - employees["in_time"]

    return (
        employees.groupby(["emp_id", "event_day"])["time_spent"]
        .sum()
        .reset_index()
        .rename(columns={"event_day": "day", "time_spent": "total_time"})
    )
```

```python
# 511. Game Play Analysis I
# https://leetcode.com/problems/game-play-analysis-i/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def game_analysis_511(activity: pd.DataFrame) -> pd.DataFrame:
    return (
        activity.groupby("player_id")
        .agg(first_login=("event_date", "min"))
        .reset_index()
    )
```

```python
# 1484. Group Sold Products By The Date
# https://leetcode.com/problems/group-sold-products-by-the-date/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def categorize_products_1484(activities: pd.DataFrame) -> pd.DataFrame:
    return (
        activities.groupby("sell_date")["product"]
        .agg(
            [
                ("num_sold", "nunique"),
                ("products", lambda x: ",".join(sorted(x.unique()))),
            ]
        )
        .reset_index()
    )
```

```python
# 1050. Actors and Directors Who Cooperated At Least Three Times
# https://leetcode.com/problems/actors-and-directors-who-cooperated-at-least-three-times/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def actors_and_directors_1050(actor_director: pd.DataFrame) -> pd.DataFrame:
    stats = (
        actor_director.groupby(
            ["actor_id", "director_id"],
        )
        .agg(
            count=("director_id", "count"),
        )
        .reset_index()
    )
    return stats[stats["count"] >= 3][["actor_id", "director_id"]]
```

```python
# 1378. Replace Employee ID With The Unique Identifier
# https://leetcode.com/problems/replace-employee-id-with-the-unique-identifier/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def replace_employee_id_1378(
    employees: pd.DataFrame, employee_uni: pd.DataFrame
) -> pd.DataFrame:
    return pd.merge(employees, employee_uni, how="left", on="id")[["unique_id", "name"]]
```

```python
# 1280. Students and Examinations
# https://leetcode.com/problems/students-and-examinations/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def students_and_examinations_1280(
    students: pd.DataFrame, subjects: pd.DataFrame, examinations: pd.DataFrame
) -> pd.DataFrame:
    return pd.merge(
        left=pd.merge(
            students,
            subjects,
            how="cross",
        ).sort_values(by=["student_id", "subject_name"]),
        right=examinations.groupby(
            ["student_id", "subject_name"],
        )
        .agg(attended_exams=("subject_name", "count"))
        .reset_index(),
        how="left",
        on=["student_id", "subject_name"],
    ).fillna(0)[["student_id", "student_name", "subject_name", "attended_exams"]]
```

```python
# 570. Managers with at Least 5 Direct Reports
# https://leetcode.com/problems/managers-with-at-least-5-direct-reports/description/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def find_managers_570(employee: pd.DataFrame) -> pd.DataFrame:
    managers = (
        employee.groupby("managerId", as_index=False)
        .agg(
            reporting=("id", "count"),
        )
        .query("5 <= reporting")["managerId"]
    )

    return employee[employee["id"].isin(managers)][["name"]]
```

```python
# 607. Sales Person
# https://leetcode.com/problems/sales-person/?envType=study-plan-v2&envId=30-days-of-pandas&lang=pythondata
def sales_person_607(
    sales_person: pd.DataFrame, company: pd.DataFrame, orders: pd.DataFrame
) -> pd.DataFrame:
    return sales_person[
        ~sales_person["sales_id"].isin(
            pd.merge(
                left=orders,
                right=company[company["name"] == "RED"],
                on="com_id",
                how="inner",
            )["sales_id"].unique()
        )
    ][["name"]]
```

```python
import pandas as pd

# Create the cust and sales DataFrames
cust = pd.DataFrame({'cust_id':[1,2,3,4,5]})
sales = pd.DataFrame({'cust_id':[2,2,2,4,4],'amount':[23,34,45,12,67]})

# Merge the two DataFrames on the cust_id column
df = cust.merge(sales.groupby('cust_id')['amount'].sum().reset_index(),how='left',on='cust_id')

# Add a count column to indicate the number of users who purchased
df['count'] = df['cust_id'].notnull().astype('int')

# Group the DataFrame by the count column and aggregate the data
result = df.groupby('count').agg({'cust_id':'count','amount':'sum'})

# Rename the columns
result.columns = ['# of users','Total amount spent']

# Print the result
print(result)
```
